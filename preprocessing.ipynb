```python
import matplotlib.pyplot as plt
import cv2
import os
from glob import glob
from tqdm import tqdm
```

```python
img_paths=glob('dataset/raw/*.png')
dataset=[]
for path in tqdm(img_paths):
    img=cv2.imread(path)
    img=cv2.cvtColor(img,cv2.COLOR_RGB2GRAY)
    _,img=cv2.threshold(img,81.2,255,cv2.THRESH_BINARY)
    dataset.append(img)
```

```python
fig,axs=plt.subplots(3,3,figsize=(10,10))
for i,ax in enumerate(axs.flat):
  ax.imshow(cv2.imread(img_paths[i]),'gray')
plt.tight_layout()
plt.show()
```

```python
fig,axs=plt.subplots(3,3,figsize=(10,10))
for i,ax in enumerate(axs.flat):
  ax.imshow(dataset[i],'gray')
plt.tight_layout()
plt.show()
```

```python
path='dataset/processed/'

for i,img in enumerate(dataset):
  cv2.imwrite(path+'{}.png'.format(i+1),img)
```
